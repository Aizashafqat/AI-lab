```python
import pandas as pd
url = "https://archive.ics.uci.edu/ml/machine-learning-databases/iris/iris.data"
df = pd.read_csv(url, header=None)  # No headers in raw Iris dataset
df.columns = ["sepal length (cm)", "sepal width (cm)", "petal length (cm)", "petal width (cm)", "species"]
```

```python
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
sns.histplot(df['sepal width (cm)'], kde=True)
plt.title('Distribution of Sepal Width')
plt.xlabel('Sepal Width (cm)')
plt.ylabel('Frequency')
plt.show()
```

```python
sns.boxplot(x='species', y='sepal length (cm)', data=df)
plt.title('Sepal Length by Species')
plt.xlabel('Species')
plt.ylabel('Sepal Length (cm)')
plt.show()
```

```python
median_sepal_width = df['sepal width (cm)'].median()
print(f"Median Sepal Width: {median_sepal_width}")
```

```python
sepal_width_exceeds_3_5 = df[df['sepal width (cm)'] > 3.5].shape[0]
print(f"Number of samples with sepal width exceeding 3.5 cm: {sepal_width_exceeds_3_5}")
```
