```python
import pandas as pd
url = "https://archive.ics.uci.edu/ml/machine-learning-databases/iris/iris.data"
df = pd.read_csv(url, header=None)  # No headers in raw Iris dataset
df.columns = ["sepal length (cm)", "sepal width (cm)", "petal length (cm)", "petal width (cm)", "species"]
```

```python
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
sns.pairplot(df[["petal length (cm)", "petal width (cm)", "sepal length (cm)", "species"]], hue="species")
plt.show()
```

```python
corr = df[["sepal length (cm)", "sepal width (cm)", "petal length (cm)", "petal width (cm)"]].corr()
sns.heatmap(corr, annot=True, cmap="coolwarm")
plt.show()
```

```python
df.groupby('species')[['petal length (cm)', 'petal width (cm)']].mean()
```

```python
df.groupby('species')[['petal length (cm)']].mean().plot(kind='bar')
plt.ylabel('petal length (cm)')
plt.title('Average Petal Length by Species')
plt.show()
```

```python
avg_petal_length = df.groupby('species')[['petal length (cm)']].mean()
highest_avg_petal_length_species = avg_petal_length['petal length (cm)'].idxmax()
maxLength = round(avg_petal_length['petal length (cm)'].max(),2)

petal_length_width_corr = df['petal length (cm)'].corr(df['petal width (cm)'])

print(f"Species with the highest average petal length: {highest_avg_petal_length_species} with length {maxLength}")
print(f"Correlation between petal length and petal width: {petal_length_width_corr}")
```
