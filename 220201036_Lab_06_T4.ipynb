```python
import pandas as pd
url = "https://archive.ics.uci.edu/ml/machine-learning-databases/iris/iris.data"
df = pd.read_csv(url, header=None)  # No headers in raw Iris dataset
df.columns = ["sepal length (cm)", "sepal width (cm)", "petal length (cm)", "petal width (cm)", "species"]
```

```python
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
species_summary = df.groupby('species')['petal width (cm)'].agg(['mean', 'median'])
print(species_summary)
```

```python
sns.boxplot(x='species', y='petal width (cm)', data=df)
plt.title('Petal Width by Species')
plt.show()
```

```python
species_summary['mean'].plot(kind='bar', title='Average Petal Width by Species')
plt.ylabel('Petal Width (cm)')
plt.show()
```

```python
highest_avg_petal_width = species_summary['mean'].idxmax()
print(f"The species with the highest average petal width is: {highest_avg_petal_width}")
```
